## 最近邻算法-kNN

### 算法描述
> 1. 计算已知类别数据集中的点与当前点之间的距离
> 2. 按照距离增序排序
> 3. 选取与当前点距离最小的k个点
> 4. 确定前k个点所在类别的出现频率
> 5. 返回前k个点出现频率最高的类别作为当前点的分类

```python
import numpy as np
import operator
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
from os import listdir
```

## 1. kNN算法实现

参数说明：
> 1. inX : 用于分类的输入向量/测试数据<br>
> 2. dataSet : 训练数据集的特征<br>
> 3. labels : 训练数据集的标签<br>
> 4. k : 选择最近邻的数目<br>
> 5. 返回值 : 输入向量的预测分类<br>

```python
def classify0(inX, dataSet, labels, k):
    
    # --------------------------------------------------------------
    # 1. 距离计算
    # dataSetSize: 训练集样本个数
    dataSetSize = dataSet.shape[0]
    
    """ diffMat: 第一行为inX坐标减去dataSet第一个点的坐标，
                 第二行为inX坐标减去dataSet第二个点的坐标，以此类推。
    """
    diffMat = np.tile(inX, (dataSetSize, 1)) - dataSet
    sqDiffMat = diffMat ** 2     # 取平方
    sqDistances = sqDiffMat.sum(axis=1)     # 每行所有元素相加
    distances = sqDistances ** 0.5      # 开方算出inX到dataSet每个点的距离
    
    # 将所有距离从小到大排序，返回对应的次序
    # argsort()将x中的元素从小到大排列，提取其对应的index（索引），然后输出到y。
    # 例如y = array([3, 0, 4, 5])，则y.argsort()=array([1, 0, 2, 3])。
    # 对应y.argsort()[0] = 1，为y[0] = 3在原数组中的次序（按增序）。
    sortedDisIndicies = distances.argsort()
    
    # ---------------------------------------------------------------
    # 2. 选取距离最小的K个点
    classCount = {}
    for i in range(k):
        voteIlabel = labels[sortedDisIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
        
    # ---------------------------------------------------------------
    # 3. 排序并返回出现最多的那个类型
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
print(classify0([0.1, 0.1], group, labels, 3))
```

在datingTestSet2.txt中主要包含了具有三个特征值的数据，样本量为1000：

> 3. 每周消费的冰淇淋公升数

首先使用file2matrix函数解析文本数据，让它变成可以被numpy处理的数据格式。

```python
def file2matrix(filename):
    """
    Parameter
    -----------
    filename : Data file path
    
    Return
    -----------
    returnMat : Features
    classLabelVector : Labels
    """
    
    fr = open(filename, 'r')
    numberOfLines = len(fr.readlines())
    returnMat = np.zeros((numberOfLines, 3))
    classLabelVector = []
    fr = open(filename, 'r')
    index = 0
    for line in fr.readlines():
        line = line.strip()
        listFromLine = line.split('\t')
        returnMat[index, :] = listFromLine[0:3]
        classLabelVector.append(int(listFromLine[-1]))
        index += 1
    return returnMat, classLabelVector
```

特征归一化<br>
> 从上面可以看到，飞行里程数的量级远远大于视频游戏时间和冰淇淋公升数，在进行距离计算的时候里程数造成的影响会远比其他两个特征要大得多。所以要进行特征的归一化处理。<br>

归一化方法：
$$Y=\frac{X-X_{min}}{X_{max}-X_{min}}$$

```python
def autoNorm(dataSet):
    """Features Noramlized
    
    Parameter
    ----------
    dataSet
    
    Returns:
    ----------
    normDataSet : DataSet after normalization
    ranges : ranges of normalization
    minVals
    """
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    
    ranges = maxVals - minVals
    normDataSet = np.zeros(dataSet.shape)
    m = dataSet.shape[0]
    normDataSet = (dataSet - minVals) / ranges
    return normDataSet, ranges, minVals
```

```python
normMat, ranges, minVals = autoNorm(datingDataMat)
normMat[0:10], ranges, minVals
```

#### 测试算法
使用90%的数据作为训练样本，10%的数据作为测试样本。通过测试样本来计算错误率。

```python
def datingClassTest():
    
    # 测试样本所占的比例
    hoRatio = 0.1
    # 加载数据
    datingDataMat, datingLabels = file2matrix("datingTestSet2.txt")
    
    # 特征归一化
    normMat, ranges, minVals = autoNorm(datingDataMat)
    m = normMat.shape[0]
    
    # 选择测试集
    numTestVecs = int(m * hoRatio)
    print('numTestVec = ', numTestVecs)
    errorCount = 0
    for i in range(numTestVecs):
        # 获取每个测试向量的标签
        classifierResult = classify0(normMat[i], normMat[numTestVecs : m],
                                     datingLabels[numTestVecs : m], 3)
        print("The classifier came back with: %d, the real answer is: %d" % 
              (classifierResult, datingLabels[i]))
        # 累加识别错误的样本数量
        errorCount += classifierResult != datingLabels[i]
    print("The total error rate is: %f" % (errorCount / numTestVecs))
    print(errorCount)
```

```python
datingClassTest()
```

#### 构建完整的可用系统

```python
def classifyPerson():
    resultList = ['not at all', 'in small doses', 'in large doses']
    percentTats = float(input("percentage of time spent playing video games?"))
    ffMiles = float(input("frequent flier miles earned per year?"))
    iceCream = float(input("liters of ice cream consumed per year?"))
    
    datingDataMat, datingLabels = file2matrix('datingTestSet2.txt')
    normMat, ranges, minVals = autoNorm(datingDataMat)
    inArr = np.array([ffMiles, percentTats, iceCream])
    classifierResult = classify0((inArr - minVals) / ranges, normMat, datingLabels, 3)
    print("You will probably like this person: %s" % (resultList[classifierResult - 1]))
```

```python
classifyPerson()
```

## 3. 手写识别系统

使用kNN算法构建手写识别系统
> 1. 收集数据
> 2. 准备数据：编写函数classify0()，将图像格式转化为分类器使用的list格式 
> 3. 分析数据
> 4. 训练算法
> 5. 测试算法：编写函数使用提供的部分数据集作为测试样本，测试样本与非测试样本的区别在于测试样本是已经完成分类的数据，如果预测分类与实际类别不同，则标记为一个错误。
> 6. 使用算法

这里的样本数据集为32x32的二进制图像矩阵，img2Vector函数将其转换为1x1024的一维向量，这样就可以使用前面的分类器处理数字图像信息了。

```python
def img2Vector(filename):
    """Transfer image data to vector
    
    Returns
    ----------
    returnVect : 1d-array of image
    
    """
    
    returnVect = np.zeros((1, 1024))
    fr = open(filename, 'r')
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0, 32 * i + j] = int(lineStr[j])
    
    return returnVect
```

```python
testVector = img2Vector('testDigits/0_13.txt')
testVector[0, 0:31]
```

```python
def handwritingClassTest():
    """Hand writing classifier"""
    
    # 1. 导入数据
    hwLabels = []
    trainingFileList = os.listdir('trainingDigits/')
    m = len(trainingFileList)
    trainingMat = np.zeros((m, 1024))
    
    for i in range(m):
        fileNameStr = trainingFileList[i]
        fileStr = fileNameStr.split('.')[0]    # 去掉每个文件的后缀名.txt
        classNumStr = int(fileStr.split('_')[0])       # 每个文件名的第一个数字为该样本的标签
        hwLabels.append(classNumStr)
        trainingMat[i] = img2Vector('trainingDigits/%s' % fileNameStr)
        
    # 2. 导入测试数据
    testFileList = os.listdir('testDigits/')
    errorCount = 0
    mTest = len(testFileList)
    for i in range(mTest):
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest = img2Vector('testDigits/%s' % fileNameStr)
        classifierResult = classify0(vectorUnderTest, trainingMat, hwLabels, 3)
        # print("The classifier came back with: %d, the real answer is: %d" % (classifierResult, classNumStr))
        errorCount += classifierResult != classNumStr
    
    print("\nThe total number of error is: %d" % errorCount)
    print("\nThe total error rate is: %f" % (errorCount / mTest))
```

```python
handwritingClassTest()
```

从这里看出，kNN算法应用在手写数据集上的错误率为1.06%。通过改变k值、修改handwritingClassTest随机选取的样本、改变训练集大小都会改变错误率。<br>
实际在使用该算法时，效率并不高。因为算法需要为每个测试向量做2000次距离计算，每个距离计算还包括了1024个维度的浮点运算。使用kd树可以使算法执行效率大幅提高。
